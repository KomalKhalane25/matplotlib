```python
#mat+plot+lib
#matrix(numpy array),plot cast,library

import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1,2,3,4,5,6])
y1 = x1**2
```

```python
print(x1)
print(y1)
```

```python
plt.plot(x1,y1)
```

```python
x1 = np.array([1,2,4,1,6])
y1 = np.array([4,2,3,8,9])
plt.plot(x1,y1)
```

```python
plt.plot(x1,y1,'red')
```

```python
plt.plot(x1,y1,'ro')
```

```python
plt.plot(x1,y1,'red')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'red',marker='*')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'r--',marker='o')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'o')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'g--')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'o--',markerfacecolor='black',markeredgecolor='green')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
#how to plot subplot
plt.subplot(2,3,1) #nrows, ncol, graph position
plt.plot(x1,y1,'r--')
plt.subplot(2,3,5)
plt.plot(y1,x1,'g*-')
plt.subplot(2,3,3)
plt.plot(x1,y1,'black')
```

```python
#scatter plot
plt.style.use('classic')
a = np.array([1,2,3,4,5])
b = np.array([4,2,3,8,9])
plt.scatter(a,b,color='red',marker='o')
```

```python
#bar plot
data= {'java':20,'C':15,'C++':30,'python':35}
courses = list(data.keys())
values = list(data.values())
plt.bar(courses,values,width=0.3,color='red')


```

```python
plt.bar(courses,values,width=0.3,color='red',edgecolor='black',linewidth=2,linestyle='--')
```

```python
fruits = ['apple','banana','strawberry','orange']
counts = [20,15,30,25]
bar_colors = ['b','yellow','r','orange']
plt.bar(fruits,counts,color=bar_colors)
```
