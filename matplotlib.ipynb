```python
#mat+plot+lib
#matrix(numpy array),plot cast,library

import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1,2,4,1,6])
y1 = np.array([4,2,3,8,9])
plt.plot(x1,y1)
```

```python
plt.plot(x1,y1,'red')
```

```python
plt.plot(x1,y1,'ro')
```

```python
plt.plot(x1,y1,'red')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'red',marker='*')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'r--',marker='o')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'o')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'g--')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
plt.plot(x1,y1,'o--',markerfacecolor='black',markeredgecolor='green')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.title('Number vs Square Plot')
```

```python
#how to plot subplot
plt.subplot(2,3,1) #nrows, ncol, graph position
plt.plot(x1,y1,'r--')
plt.subplot(2,3,5)
plt.plot(y1,x1,'g*-')
plt.subplot(2,3,3)
plt.plot(x1,y1,'black')
```

```python
#scatter plot
plt.style.use('classic')
a = np.array([1,2,3,4,5])
b = np.array([4,2,3,8,9])
plt.scatter(a,b,color='red',marker='o')
```

```python
#bar plot
data= {'java':20,'C':15,'C++':30,'python':35}
courses = list(data.keys())
values = list(data.values())
plt.bar(courses,values,width=0.3,color='red')


```

```python
plt.bar(courses,values,width=0.3,color='red',edgecolor='black',linewidth=2,linestyle='--')
```

```python
fruits = ['apple','banana','strawberry','orange']
counts = [20,15,30,25]
bar_colors = ['b','yellow','r','orange']
plt.bar(fruits,counts,color=bar_colors)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.5,0.1,0.8,0.6]) #left,bottom,wigth,height,range(0,1)
x1= np.array([1,2,3,4,5,6])
y1 = x1**2
ax.plot(x1,y1)
ax.set_xlabel('Number')
ax.set_ylabel('Square of Number')
ax.set_title('Number vs Square Plot')
```

```python
fig = plt.figure()
axexs1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,wigth,height,range(0,1)
axexs2 = fig.add_axes([0.2,0.5,0.4,0.3])

axexs1.plot(x1,y1)
axexs1.set_xlabel('Numbers axes1')
axexs1.set_ylabel('Square of Number1')
axexs2.plot(y1,x1,'red')
axexs2.set_xlabel('Number')
axexs2.set_ylabel('Square of Number2')
axexs1.set_ylabel('Numbers axes2')
```

```python
fig = plt.figure()
axexs1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,wigth,height,range(0,1)
axexs2 = fig.add_axes([0.1,0.3,0.5,0.4])

axexs1.plot(x1,y1)
axexs1.set_xlabel('Numbers axes1')
axexs1.set_ylabel('Square of Number1')
axexs2.plot(y1,x1,'red')
axexs2.set_xlabel('Number')
axexs2.set_ylabel('Square of Number2')
axexs1.set_ylabel('Numbers axes2')
```

```python
fig = plt.figure(figsize=(10,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x1,x1**2,label='square')
ax.plot(x1,x1**3,label='cube')
ax.legend()

```
